```python
import daft
from daft import PGM
```

```python
pgm = daft.PGM(shape=[3.5, 2.5], origin=[0, 0], grid_unit=4,
               label_params={'fontsize':18}, observed_style='shaded')


pgm.add_node(daft.Node("beta", r"$\beta$", 1, 2, scale=2))
pgm.add_node(daft.Node("X", r"$X_i$", 2, 2, scale=2, observed=True))
pgm.add_node(daft.Node("r", r"$r_i$", 1, 1, scale=2, observed=True))

pgm.add_plate(daft.Plate([0.5, 0.7, 2., 1.8], shift=-0.1))
pgm._ctx.ax().annotate(r"$t = 1, \ldots, T$", [7.2, 2.75], fontsize=14)

pgm.add_edge("beta", "r")
pgm.add_edge("X", "r")

pgm.render();
# pgm.figure.savefig("images/simplest-logistic-response-model-pgm.png", dpi=300)
```
