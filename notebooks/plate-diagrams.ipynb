```python
import daft
from daft import PGM
```

```python
pgm = daft.PGM(shape=[3.5, 2.5], origin=[0, 0], grid_unit=4,
               label_params={'fontsize':18}, observed_style='shaded')


pgm.add_node(daft.Node("beta", r"$\beta$", 1, 2, scale=2))
pgm.add_node(daft.Node("X", r"$X_i$", 2, 1, scale=2, observed=True))
pgm.add_node(daft.Node("r", r"$r_i$", 1, 1, scale=2, observed=True))

pgm.add_plate(daft.Plate([0.5, 0.5, 2, 1.0], shift=-0.1))
pgm._ctx.ax().annotate(r"$i = 1, \ldots, t$", [2.1, 1.75])

pgm.add_edge("beta", "r")
pgm.add_edge("X", "r")

pgm.render();
# pgm.figure.savefig("images/simplest-logistic-response-model-pgm.png", dpi=300)
```

```python
pgm = daft.PGM(shape=[3.5, 3.5], origin=[0, 0], grid_unit=4,
               label_params={'fontsize':18}, observed_style='shaded')


pgm.add_node(daft.Node("beta", r"$\beta$", 1, 3, scale=2))
pgm.add_node(daft.Node("X", r"$X_i$", 2, 2, scale=2, observed=True))
pgm.add_node(daft.Node("omega", r"$\omega_i$", 1, 2, scale=2, observed=False))
pgm.add_node(daft.Node("r", r"$r_i$", 1, 1, scale=2, observed=True))

daft.Plate
pgm.add_plate(daft.Plate([0.5, 0.5, 2, 2.], shift=-0.1))
pgm._ctx.ax().annotate(r"$i = 1, \ldots, t$", [2.1, 1.75])

pgm.add_edge("beta", "omega")
pgm.add_edge("X", "omega")
pgm.add_edge("X", "r")
pgm.add_edge("omega", "r")

pgm.render();
# pgm.figure.savefig("images/pg-logistic-response-model-fixed-beta-pgm.png", dpi=300)
```

```python
pgm = daft.PGM(shape=[3.5, 4.5], origin=[0, 0], grid_unit=4,
               label_params={'fontsize':18}, observed_style='shaded')

pgm.add_node(daft.Node("mu", r"$\mu$", 1, 4, scale=2, fixed=True))
pgm.add_node(daft.Node("Sigma", r"$\Sigma$", 2, 4, scale=2, fixed=True))
pgm.add_node(daft.Node("beta", r"$\beta$", 1, 3, scale=2))
pgm.add_node(daft.Node("X", r"$X_i$", 2, 2, scale=2, observed=True))
pgm.add_node(daft.Node("omega", r"$\omega_i$", 1, 2, scale=2, observed=False))
pgm.add_node(daft.Node("r", r"$r_i$", 1, 1, scale=2, observed=True))

daft.Plate
pgm.add_plate(daft.Plate([0.5, 0.5, 2, 2.], shift=-0.1))
pgm._ctx.ax().annotate(r"$i = 1, \ldots, t$", [2.1, 1.75])

pgm.add_edge("mu", "beta")
pgm.add_edge("Sigma", "beta")
pgm.add_edge("beta", "omega")
pgm.add_edge("X", "omega")
pgm.add_edge("X", "r")
pgm.add_edge("omega", "r")

pgm.render();
# pgm.figure.savefig("images/pg-logistic-response-model-pgm.png", dpi=300)
```

```python
pgm = daft.PGM(shape=[3.5, 5.5], origin=[0, 0], grid_unit=4,
               label_params={'fontsize':18}, observed_style='shaded')

# Hyperparameters
pgm.add_node(daft.Node("Lambda", r"$\Lambda_0$", 2.7, 4.5, scale=2, fixed=True))
pgm.add_node(daft.Node("nu", r"$\nu_0$", 2, 4.5, scale=2, fixed=True))
pgm.add_node(daft.Node("mu0", r"$\mu_0$", 1, 4.5, scale=2, fixed=True))
pgm.add_node(daft.Node("kappa0", r"$\kappa_0$", 0.3, 4.5, scale=2, fixed=True))

# Parameters
pgm.add_node(daft.Node("mu", r"$\mu$", 1, 4, scale=2))
pgm.add_node(daft.Node("Sigma", r"$\Sigma$", 2, 4, scale=2))
pgm.add_node(daft.Node("beta", r"$\beta$", 1, 3, scale=2))
pgm.add_node(daft.Node("X", r"$X_i$", 2, 2, scale=2, observed=True))
pgm.add_node(daft.Node("omega", r"$\omega_i$", 1, 2, scale=2, observed=False))
pgm.add_node(daft.Node("r", r"$r_i$", 1, 1, scale=2, observed=True))

# Data instance plate
daft.Plate
pgm.add_plate(daft.Plate([0.5, 0.5, 2, 2.], shift=-0.1))
pgm._ctx.ax().annotate(r"$i = 1, \ldots, t$", [2.1, 1.75], fontsize=14)

pgm.add_edge("Lambda", "Sigma")
pgm.add_edge("nu", "Sigma")
pgm.add_edge("mu0", "mu")
pgm.add_edge("kappa0", "mu")
pgm.add_edge("mu", "beta")
pgm.add_edge("Sigma", "beta")
pgm.add_edge("Sigma", "mu")
pgm.add_edge("beta", "omega")
pgm.add_edge("X", "omega")
pgm.add_edge("X", "r")
pgm.add_edge("omega", "r")

pgm.render();
# pgm.figure.savefig("images/bayes-pg-logistic-response-model-pgm.png", dpi=300)
```
